```python
#importing the libraries
import numpy as np
import torch
```

```python
#creating 2 tensors
t1 = torch.tensor([[1,2],[3,4]])
t2 = torch.tensor([[5,6],[7,8]])
```

```python
#Element wise operations of tensor
print("Tensor addition : {}\nTensor Subtraction : {}\nTensor Multiplication : {}\nTensor Division : {}".format(t1+t2,t1-t2,t1*t2,t1//t2))
```

```python
#broadcasting with scalars
print("{}\n{}".format(t1 + 2,t1 - 2))
```

# BroadCasting

broadcasting is an feature that allows us to perform arithmetic operations on tensors of different sizes

In the above example, the scalar "2" is converted into a tensor of t1's shape and then the addition and subtraction is performed 

```python
#broadcasting tensors
t3 = torch.tensor([[1,2],[3,4]])
t5 = torch.tensor([5,5])
print(t3+t5)
```

BroadCasting can also be done on multiple tensors.

In the above example , a row is appended to t5 with the same values of the first row. then, addition is performed

# Reduction operations

Reduction operations are operations that when performed, reduce the size of the tensor

```python
#reduction operations 
print(t1)
print("Some of the reduction operations are:\nadd:{}\nmax:{}\nproduct:{}".format(t1.sum(),t1.max(),t1.prod()))
```

The argmax function flattens a given tensor and then finds the index of the maximum element in tensor

```python
#argmax function
print(t1,"\nindex of the largest element is : ",t1.argmax())
```

We can convert the pytorch tensors to regular python datatypes by using item adn tolist functions.
Here's an Example

```python
#tensor to python datatypes
print(t1.max().item(),"\n{}".format(t1.flatten().tolist()))
```
