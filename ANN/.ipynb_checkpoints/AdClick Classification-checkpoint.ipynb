```python
#importing the libraries
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data

The dataset is available at kaggle : https://www.kaggle.com/dragonheir/logistic-regression

```python
#importing the dataset
df = pd.read_csv('Social_Network_Ads.csv')
df.head()
```

removing User Id as it has no relationship with ad clicks

```python
#droping insignificant columns
df.drop(['User ID'],inplace=True,axis=1)
```

```python
#removing the gender column as there is not much effect on the purchased
df.drop(['Gender'],axis=1,inplace=True)
```

```python
df.head()
```

```python
x = np.array(df.iloc[:,:2])
y = np.array(df.iloc[:,2])
```

# Feature Scaling

scaling the data is very important for gradient descent to work

```python
print('shape of X : ',x.shape)
print('Shape of Y : ',y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
print('shape of the x_train: ',x_train.shape)
print('shape of the x_test : ',x_test.shape)
print('shape of the y_train: ',y_train.shape)
print('shape of the x_test: ',y_test.shape)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Dataset

PyTorch dataset class makes it easier for accessing elements of the tensor

```python
#dataset class
from torch.utils.data import Dataset
class dataset(Dataset):
  def __init__(self,x,y):
    self.x = torch.tensor(x,dtype=torch.float32)
    self.y = torch.tensor(y,dtype=torch.float32)
    self.length = self.x.shape[0]
  
  def __getitem__(self,idx):
    return self.x[idx],self.y[idx]

  def __len__(self):
    return self.length
```

```python
#training and testing tensors
train_set = dataset(x_train,y_train)
test_set = dataset(x_test,y_test)
```

# Neural Network

It's a simple neural network with sigmoid activation at the final layer

```python
#defining the network
from torch import nn
from torch.nn import functional as F

class LogisticRegression(nn.Module):
  def __init__(self):
    super(LogisticRegression,self).__init__()
    self.fc1 = nn.Linear(2,1)

  def forward(self,x):
    x = torch.sigmoid(self.fc1(x))
    return x
```

```python
#some parameters
model = LogisticRegression()
learning_rate = 0.007
optimizer = torch.optim.Adam(model.parameters(),lr=learning_rate)
epochs = 1000
criterion = nn.BCELoss()
```

# DataLoader

dataloader makes things easier for the forward pass and in this case, I have performed batch gradient descent by specifying the batch size to be equal to size of the training set

```python
#data loader
from torch.utils.data import DataLoader
train_loader = DataLoader(train_set,shuffle=True,batch_size=train_set.__len__())
test_loader = DataLoader(test_set,batch_size=test_set.__len__())
```

# Forward and Backward Pass

predict, get the loss, calculate the gradients with respect to the loss, update weights

```python
#forward pass
losses = []
for i in range(epochs):
  for j,(x_train,y_train) in enumerate(train_loader):
    #get the prediction 
    y_pred = model(x_train)

    #losses
    loss = criterion(y_pred,y_train.reshape(-1,1))
    losses.append(loss)
    #backprop
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

  #print loss
  if i%100 == 0:
    print("epoch : {} loss: {}".format(i,loss))
```

# Model Analysis

plotting loss over epochs in training set and calculating accuracy of the model's prediction of the test set

confusion matrix is also displayed below

```python
plt.plot(losses)
plt.title('loss vs epochs')
plt.xlabel('epochs')
plt.ylabel('loss')
```

```python
#testset
x_test,y_test = next(iter(train_loader))
y_pred = model(x_test)
print('accuracy of the model on test set : ',(((y_pred.round().reshape(-1) == y_test).sum())/float(y_pred.shape[0])).item(),"%")
```

# Confusion Matrix

It returns 4 values

1st Value represents : True Positive

2nd Value Represents : False Positive

3rd Value represents : False Negative 

4th Value represents : True Negative

confusion matrix.png

```python
from sklearn.metrics import confusion_matrix
print("Confusion Matrix : ")
confusion_matrix(y_pred.round().reshape(-1).detach(),y_test)
```
